```python
def printSubsequences(arr, index, subarr):
       
    # Print the subsequence when reach 
    # the leaf of recursion tree
    if index == len(arr):
           
        # Condition to avoid printing
        # empty subsequence
        if len(subarr) != 0:
            print(subarr)
       
    else:
        # Subsequence without including 
        # the element at current index
        printSubsequences(arr, index + 1, subarr)
           
        # Subsequence including the element
        # at current index
        printSubsequences(arr, index + 1, 
                            subarr+[arr[index]])
       
    return
           
arr=[1,2,3]

   
printSubsequences(arr,0,[])
```

```python
# Recursive Python program for insertion sort

# Recursive function to sort an array using insertion sort
def insertionSortRecursive(arr, n):
	# base case
	if n <= 1:
		return

	# Sort first n-1 elements
	insertionSortRecursive(arr, n - 1)

	# Insert last element at its correct position in sorted array.
	last = arr[n - 1]
	j = n - 2

	# Move elements of arr[0..i-1], that are
	# greater than key, to one position ahead
	# of their current position
	while (j >= 0 and arr[j] > last):
		arr[j + 1] = arr[j]
		j = j - 1
	arr[j + 1] = last


# Driver program to test insertion sort
if __name__ == '__main__':
	A = [-7, 11, 6, 0, -3, 5, 10, 2]
	n = len(A)
	insertionSortRecursive(A, n)
	print(A)

# Contributed by Harsh Valecha,
# Edited by Abraar Masud Nafiz.
```

```python
from functools import cmp_to_key
class Employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def printDetails(self):
        return {'name':self.name,'salary':self.salary} 
    def comperator(a,b):
        if a.salary>b.salary:
            return -1
        if a.salary<b.salary:
            return 1
        if a.name<b.name:
            return -1
        if a.name>b.name:
            return 1
        return 0 
n=int(input())
arr=[]
for i in range(n):
    name,salary=input().split()
    salary=float(salary)
    Emp=Employee(name,salary)
    arr.append(Emp)
arr=sorted(arr,key=cmp_to_key(Employee.comperator))
for i in arr:
    print(i.name,i.salary)    
```
